### errors : ‘ignore’, ‘raise’, ‘coerce’, default ‘raise’

* 1. If ‘raise’, then invalid parsing will raise an exception.
* 2. If ‘coerce’, then invalid parsing will be set as NaN.
* 3. If ‘ignore’, then invalid parsing will return the input.

 In our case, we will use errors = ‘coerce’ to convert any non-numeric values to null NaN.

```python
autos_df['horsepower']
```

```python
autos_df['horsepower'] = pd.to_numeric(autos_df['horsepower'],errors='coerce')
```

```python
autos_df.info()
```

```python
autos_df['horsepower']
```

```python
autos_df[autos_df.horsepower.isnull()]
```

```python
autos_df = autos_df.dropna(subset=['horsepower'])
```

```python
autos_df[autos_df.horsepower.isnull()]
```

### Box Plot 

Box plot (aka Box and Whisker plot) is a graphical representation of numerical data that can be used to 
understand the variability of the data and the existence of outliers. 

1. Lower quartile (1st quartile), median and upper quartile (3rd quartile).
2. Lowest and highest values.
3. Inter-quartile range (IQR).



```python
box = sns.boxplot(df['SOLD PRICE'])
```

```python
box = plt.boxplot(df['SOLD PRICE'])

## boxplot with median line
```

```python
[item.get_ydata()[0] for item in box['caps']]

"""
The output [np.int64(20000), np.int64(1350000)] represents the Whiskers' Range:

20,000: The Minimum value (excluding outliers). Anything below this is considered a "bottom outlier."

1,350,000: The Maximum value (excluding outliers). Anything above this is a "top outlier."
"""
```

```python
[item.get_ydata()[0] for item in box['whiskers']]
## The boundaries of the box (Q1 and Q3).
```


It is a way to "group" or "categorize" your data points visually.

* The Simple Analogy
Imagine you have a list of students' Height and Weight.

If you plot them on a graph, you just see a bunch of dots.

If you add hue='Gender', the dots will turn Blue for Boys and Pink for Girls.

Now, you can instantly see if boys are generally taller than girls. The color (hue) told you the secret story.

* A Real-World Example
Suppose you are looking at House Prices based on Square Footage.

Without Hue: You see that as size increases, price increases. (A simple scatter plot).

With hue='City': The dots are colored differently for Mumbai, Delhi, and Guwahati.

Suddenly, you notice that a small house in Mumbai (Red dots) costs as much as a big house in Guwahati (Green dots). The hue helped you compare categories.


```python
sns.histplot(df[df['CAPTAINCY EXP']==1]
             ['SOLD PRICE'],
             color = 'y',
             label = 'Captaincy Experience')

sns.histplot(df[df['CAPTAINCY EXP']==0]
             ['SOLD PRICE'],
             color='r',
             label = 'No Captaincy Experience')

plt.legend()
```

```python
## Another simple way to do this
## This automatically splits the data, assigns colors, and adds a legend
sns.histplot(data=df, x='SOLD PRICE', hue='CAPTAINCY EXP', palette={'y', 'r'})
```
