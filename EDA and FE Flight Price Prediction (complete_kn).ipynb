## Basics.

1) When we use [] (square brackets)
We use square brackets to select something from a DataFrame.
Eg: 
df['Age']     # selecting one column
df[['Age','City']]   # selecting multiple columns
df[ df['Age'] > 20 ] # selecting rows based on condition

You are basically saying:
“From this DataFrame, give me this part.”

2) When we use () (parentheses)
We use parentheses to run a function or call a method.
Eg: df.head()      # show first rows
df.info()      # show summary
df.describe()  # calculate statistics
df['Age'].mean()  # calculate mean of Age

You are basically saying:
“Do this action.”

Eg : df['Salary'].max()

df['Salary'] → selecting the Salary column ([])
.max() → calculating maximum value (())

So:

First we select → then we act.

#### EDA (Exploratory Data Analysis)

Goal: Understand the data.

In EDA, we observe, we do not change anything.

We are just studying the data.

What we do in EDA:

Load the dataset

Check shape (how many rows and columns)

Check data types (numeric or categorical)

Check missing values (NaN values)

Check duplicated rows

Check unique categories

Summary statistics (mean, median, min, max, std)

Visualizations

Bar plots (Category Counts)

Histograms/KDE (Distribution of numeric columns)

Boxplots (Outliers)

Heatmap (Correlation between numerical values)

Scatter Plot (Relationship between two variables)

In one line:

EDA = Understanding data patterns, trends, and problems.

You don’t modify the data here.
You only explore.

#### 🛠 Feature Engineering (FE)

Goal: Improve data for model / analysis.

Here, we transform the data.
What we do in Feature Engineering:

Handle missing values

Fill mean/median/mode

Or remove rows

Handle duplicates

Remove repeated rows

Convert data types

Like '3.0M' → 3000000

Encode categorical columns

Label Encoding

One Hot Encoding

Scale/Normalize numeric values

StandardScaler / MinMaxScaler

Create new useful features

Example: Convert Size → Size_Display

Extract month from date

Convert text length into number

Remove unnecessary columns

Drop ID, Unnamed columns, etc.

In one line:

FE = Cleaning + transforming data to make it ready for model training.

You modify the data here.

### Let's Do FE and EDA Parallely¶

```python
## Feature engg process we are going to start

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]

"""
⭐ Why is this done? (Very Important)

Machine Learning models cannot understand date formats like "24/03/2019".
"""
## This splits "24/03/2019" into a list:
## ["24", "03", "2019"]
# .str[0]	First part	Date
# .str[1]	Second part	Month
# .str[2]	Third part	Year
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])

"""
lambda x: x.split(' ')[0]
.split(' ') → splits the string wherever there is a space

eg. 01:10 22 Mar
    "22:45 13 Mar" → ["22:45", "13", "Mar"]
    "19:00 Next Day" → ["19:00", "Next", "Day"]

So final result:

Before (Arrival_Time)	After
"22:45 13 Mar"	       "22:45"
"05:25 14 Mar"	       "05:25"
"19:00 Next Day"	   "19:00"
"""

## To seperate date from arrival time eg. 01:10 22 Mar
```

```python
df.head(2)
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min']=df['Arrival_Time'].str.split(':').str[1]

## if we have '/' then str.split('/') and if we have ':' then str.split(':').
```

```python
df.head(2)
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Min'] = df['Arrival_Min'].astype(int)
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```
